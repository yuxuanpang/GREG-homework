```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
import pandas as pd  

log_T=np.arange(2.8,8.2,0.01)

def fun1(T):
    ka1=np.log10(0.20)+np.log10(1+0.70)+0*T
    return ka1

def fun2(T):
    ka2=np.log10(3.8*10**22)+np.log10(1+0.70)-6-3.5*T
    return ka2

def fun3(T):
    ka3=np.log10(4.3*10**25)+np.log10(1+0.70)-6-3.5*T+np.log10(0.02)
    return ka3

def fun4(T):
    ka4=np.log10(2.5*10**(-31))-3+9*T+np.log10(0.02/0.02)
    return ka4
```

```python
#1
ka1=fun1(log_T)
ka2=fun2(log_T)

#2
ka3=fun3(log_T)

#3
ka4=fun4(log_T)

ka_all1=ka4[log_T<=3.8]
ka_all2=[]
for each in range(len(log_T)):
    if log_T[each]>3.8 and log_T[each]<4.2:
        ka_all2.append(ka4[each]*(1-2.5*(log_T[each]-3.8))+2.5*(log_T[each]-3.8)*np.log10(10**ka1[each]+10**ka2[each]+10**ka3[each]))
ka_all3=np.log10(10**ka1[log_T>4.2]+10**ka2[log_T>4.2]+10**ka3[log_T>4.2])
ka_all4=np.append(ka_all1,ka_all2,axis=0)
ka_all=np.append(ka_all4,ka_all3,axis=0)


font1 = {'family' : 'Times New Roman','weight' : 'normal','size' : 18,}
f, ax = plt.subplots(figsize = (15,10))
plt.tick_params(labelsize=18)
ax.set_title('The opacity curve',fontsize=30)
ax.set_xlabel(r'$\log_{10}T[g/cm^3]$',fontsize=24)
ax.set_ylabel(r'$\log_{10}\kappa$',fontsize=24)
plt.plot(log_T[log_T>4],ka1[log_T>4],'--',label='electron scattering')
plt.plot(log_T[log_T>4],ka2[log_T>4],'--',label='free-free absortion')
plt.plot(log_T[log_T>4],ka3[log_T>4],'--',label='bound-free absortion')
plt.plot(log_T[log_T<=4],ka4[log_T<=4],'--',label='the H iron')
plt.plot(log_T,ka_all,label='the opacity curve')
plt.axis([2.8,8.2,-4.5,5.5])
legend = plt.legend(prop=font1)
plt.savefig('Q3-5-1.jpeg')
plt.show()
```
