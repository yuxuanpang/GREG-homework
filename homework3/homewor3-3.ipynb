```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
import pandas as pd  

log_rho=np.arange(-7,13,0.01)
```

```python
e=1.6*10**(-19)
m_u=1.674*10**(-27)
k=1.38*10**(-23)
mol=6.02*10**23
mu=0.5
mu_e=1
pi=3.1416
h=6.62*10**(-34)
m_e=9.1*10**(-31)
c=3*10**8

C1=1/mu*k/m_u#ideal gas
C2=h*h/20/m_e*(3/pi)**(2/3)/mu_e**(5/3)/m_u**(5/3)#non-rel degeneracy
C3=h*c/8*(3/pi)**(1/3)/mu_e**(4/3)/m_u**(4/3)#rel degeneracy
C4=8*pi**5*k**4/45/h**3/c**3#radiation
#C5=e*e/k*(4*pi/3/m_u)**(1/3)
C5=2.275*10**5
def fun1(rho):
    T1=-np.log10(C1)+np.log10(C2)+2/3*rho
    return T1

def fun2(rho):
    T2=-np.log10(C1)+np.log10(C3)+1/3*rho
    return T2

def fun3(rho):
    T3=1/3*(-np.log10(C4)+np.log10(C1)+rho)
    return T3

def fun4(rho):
    T4=1/3*rho+np.log10(C5)-np.log10(170)
    return T4

A3=pi/3*m_u*mu_e*(5*m_e*c/2/h)**3
rho_r=np.log10(A3)
```

```python
#1
T1=fun1(log_rho)
T2=fun2(log_rho)

#2
T3=fun3(log_rho)

#3
T4=fun4(log_rho)

#4
f, ax = plt.subplots(figsize = (15,10))
#set the size of x and y axis
plt.tick_params(labelsize=18)
ax.set_title('Contours',fontsize=30)
#ax.set_xlabel('$log_{10}\rho$',fontsize=24)
#ax.set_ylabel('$log_{10}T$',fontsize=24)
plt.plot(log_rho[log_rho<rho_r]-3,T1[log_rho<rho_r])
plt.plot(log_rho[log_rho>rho_r]-3,T2[log_rho>rho_r])
plt.plot(log_rho-3,T3)
plt.vlines(rho_r-3,-2,fun2(rho_r))
plt.plot(log_rho-3,T4)
plt.annotate('Ideal gas',(-7.8,3.3), family='Times New Roman',size=40,color='black')
plt.annotate('Radiation',(-7.8,10), family='Times New Roman',size=40,color='black')
plt.annotate('NR',(4,2.3), family='Times New Roman',size=30,color='black')
plt.annotate('ER',(8,2.3), family='Times New Roman',size=30,color='black')
plt.annotate('Degenerate eletron',(4,6), family='Times New Roman',size=30,color='black',rotation='30')
plt.annotate('Crystalliztion',(3,4), family='Times New Roman',size=30,color='black',rotation='25')
plt.axis([-10, 10, 2, 12])
```
